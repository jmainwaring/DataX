```python
original_df = pd.read_csv("Restaurant.csv")
```

```python
original_df.head()
```

```python
original_df.info()
```

```python
# Creating a copy of the data frame
df = original_df.copy()
```

```python
# This converts all columns with "object" variables (AKA string) into numbers, and creates a dictionary  
char_cols = df.dtypes.pipe(lambda x: x[x == 'object']).index
label_mapping = {}

for c in char_cols:
    df[c], label_mapping[c] = pd.factorize(df[c])
```

```python
# Here you can see the order it assigned the ints to string. In the City column, Istanbul=0, Ankara=1, etc.
label_mapping
```

```python
# If you want to access the mapping for one specific column you can do it this way
label_mapping["City"]
```

```python
# List of random numbers (no repeats) between 1 and 137 and then delete A_follower_count[that row]   
rows_to_delete = random.sample(range(137), 30)
```

```python
# Deleting values from those rows
for x in rows_to_delete:
    df['revenue'][x] = np.nan
```

```python
original_df.head(20)
```

```python
# Creating array of the correct answers from original data frame  
deleted_answer_list = []
for x in rows_to_delete:
    deleted_answer_list.append(original_df['revenue'][x])

# Converting it to array from a list so we can perform certain calculations
deleted_answer_array = np.array(deleted_answer_list)
```

```python
deleted_answer_array
```

```python
# Making sure the 30 values were removed from that column (you can see there are only 107 values in 
# the 'revenue' column now)
df.info()
```

```python
# Need to first split the data into complete/not-complete, then split complete into target/features, train
# it, then run that on the empty ones 
```

```python
# Splitting into complete and non-complete
```

```python
# Accessing the 107 rows without empty "revenue" values
df_complete_revenue = df.dropna()
df_complete_revenue.shape
```

```python
# Accessing the 30 rows with empty "revenue" values
df_empty_revenue = df.loc[df['revenue'].isnull()]
df_empty_revenue.shape
```

```python
# Splitting complete rows into target/features
```

```python
features = df_complete_revenue.drop('revenue', axis=1)
```

```python
features.head()
```

```python
target_variable = df_complete_revenue['revenue']
```

```python
target_variable.head()
```

```python
# Splitting the rows with empty "revenue" into features and response (which is what we're predicting)
features_empty = df_empty_revenue.drop('revenue', axis=1)
target_variable_empty = df_empty_revenue.drop('revenue', axis=1)
```

```python
# Training ridge and lasso models (we should use ridge since we're less concerned with fully removing variables)
```

```python
from sklearn.linear_model import Ridge
from sklearn.utils import shuffle
from sklearn.linear_model import Lasso
```

```python
# Ridge Regression
ridgereg = Ridge(normalize=True)
ridgereg.fit(features,target_variable)
y_pred_ridge = ridgereg.predict(features_empty)
print ("RMSE: " + str((mean_squared_error(deleted_answer_array, y_pred_ridge))**0.5))
```

```python
# Lasso Regression
lassoreg = Lasso(normalize=True)
lassoreg.fit(features,target_variable)
y_pred_lasso = lassoreg.predict(features_empty)
print ("RMSE: " + str((mean_squared_error(deleted_answer_array, y_pred_lasso))**0.5))
```

```python
# Random forests
```

```python
df['City Group'].describe()
```

```python
# Creating another copy of the dataframe
df2 = original_df.copy()
```

```python
# This converts all columns with "object" variables (AKA string) into numbers, and creates a dictionary  
char_cols = df2.dtypes.pipe(lambda x: x[x == 'object']).index
label_mapping = {}

for c in char_cols:
    df2[c], label_mapping[c] = pd.factorize(df2[c])
```

```python
# List of random numbers (no repeats) between 1 and 137 and then delete A_follower_count[that row]   
rows_to_delete = random.sample(range(137), 30)
```

```python
# Creating array of the correct answers from original data frame  
deleted_answer_list = []
for x in rows_to_delete:
    deleted_answer_list.append(df2['City Group'][x])

# Converting it to array from a list so we can perform certain calculations
deleted_answer_array_city_group = np.array(deleted_answer_list)
deleted_answer_array_city_group
```

```python
# Deleting values from those rows
for x in rows_to_delete:
    df2['City Group'][x] = np.nan
```

```python
df2.head(8)
```

```python
# Splitting into complete and non-complete
```

```python
# Accessing the 107 rows without empty "revenue" values
df_complete_city_group = df2.dropna()
df_complete_city_group.shape
```

```python
# Accessing the 30 rows with empty "revenue" values
df_empty_city_group = df2.loc[df2['City Group'].isnull()]
df_empty_city_group.shape
```

```python
# Splitting into features (everything except "City Group" column) and target variable ("City Group" column)
```

```python
# Features for training set
features_city_group = df_complete_city_group.drop('City Group', axis=1)
features_city_group.head()
```

```python
# Target for training set 
target_variable_city_group = df_complete_city_group['City Group']
target_variable_city_group.head()
```

```python
# Features for test set
features_empty_city_group = df_empty_city_group.drop('City Group', axis=1)
features_empty_city_group.head()
```

```python
# Training random forest
randFor = RandomForestClassifier(n_estimators = 20)
randFor.fit(features_city_group, target_variable_city_group)
```

```python
# Accuracy on the training set set
randFor.score(features_city_group, target_variable_city_group)
```

```python
# Set of "City Group" predictions for the rows with empty values 
y_pred_randFor = randFor.predict(features_empty_city_group)
```

```python
# Accuracy on the test set
accuracy_score(y_pred_randFor, deleted_answer_array_city_group)
```

```python
# Trying the same thing but with a gradient boosting classifier
```

```python
gradBoost = GradientBoostingClassifier()
gradBoost.fit(features_city_group, target_variable_city_group)
```

```python
# Accuracy on the training set set
gradBoost.score(features_city_group, target_variable_city_group)
```

```python
# Set of "City Group" predictions for the rows with empty values 
y_pred_gradBoost = gradBoost.predict(features_empty_city_group)
```

```python
# Accuracy on the test set
accuracy_score(y_pred_gradBoost, deleted_answer_array_city_group)
```

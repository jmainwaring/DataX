```python
original_df = pd.read_csv("Restaurant.csv")
```

```python
original_df.head()
```

```python
original_df.info()
```

```python
# Creating a copy of the data frame
df = original_df.copy()
```

```python
# This converts all columns with "object" variables (AKA string) into numbers, and creates a dictionary  
char_cols = df.dtypes.pipe(lambda x: x[x == 'object']).index
label_mapping = {}

for c in char_cols:
    df[c], label_mapping[c] = pd.factorize(df[c])
```

```python
# Here you can see the order it assigned the ints to string. In the City column, Istanbul=0, Ankara=1, etc.
label_mapping
```

```python
# If you want to access the mapping for one specific column you can do it this way
label_mapping["City"]
```

```python
# List of random numbers (no repeats) between 1 and 137 and then delete A_follower_count[that row]   
rows_to_delete = random.sample(range(137), 30)
```

```python
# Deleting values from those rows
for x in rows_to_delete:
    df['revenue'][x] = np.nan
```

```python
original_df.head(20)
```

```python
# Creating array of the correct answers from original data frame  
deleted_answer_list = []
for x in rows_to_delete:
    deleted_answer_list.append(original_df['revenue'][x])

# Converting it to array from a list so we can perform certain calculations
deleted_answer_array = np.array(deleted_answer_list)
```

```python
deleted_answer_array
```

```python
# Making sure the 30 values were removed from that column (you can see there are only 107 values in 
# the 'revenue' column now)
df.info()
```

```python
# Need to first split the data into complete/not-complete, then split complete into target/features, train
# it, then run that on the empty ones 
```

```python
# Splitting into complete and non-complete
```

```python
# Accessing the 107 rows without empty "revenue" values
df_complete_revenue = df.dropna()
df_complete_revenue.shape
```

```python
# Accessing the 30 rows with empty "revenue" values
df_empty_revenue = df.loc[df['revenue'].isnull()]
df_empty_revenue.shape
```

```python
# Splitting complete rows into target/features
```

```python
features = df_complete_revenue.drop('revenue', axis=1)
```

```python
features.head()
```

```python
target_variable = df_complete_revenue['revenue']
```

```python
target_variable.head()
```

```python
# Splitting the rows with empty "revenue" into features and response (which is what we're predicting)
features_empty = df_empty_revenue.drop('revenue', axis=1)
target_variable_empty = df_empty_revenue.drop('revenue', axis=1)
```

```python
# Training various models - the type of model we use will be dependent on if the value is categorical or not...
```

```python
linreg_follower_count = LinearRegression()
linreg_follower_count.fit(features, target_variable)

linreg_score=linreg_follower_count.score(features, target_variable)
print ('R_squared:',linreg_score)
```

```python
# Try out lasso
```

```python
from sklearn.linear_model import Ridge
from sklearn.utils import shuffle
from sklearn.linear_model import Lasso
```

```python
# Ridge Regression
ridgereg = Ridge(normalize=True)
ridgereg.fit(features,target_variable)
y_pred_ridge = ridgereg.predict(features_empty)
print ("RMSE: " + str((mean_squared_error(deleted_answer_array, y_pred_ridge))**0.5))
```

```python
# Lasso Regression
lassoreg = Lasso(normalize=True)
lassoreg.fit(features,target_variable)
y_pred_lasso = lassoreg.predict(features_empty)
print ("RMSE: " + str((mean_squared_error(deleted_answer_array, y_pred_lasso))**0.5))
```
